```python
import pandas as pd
import networkx as nx
```

```python
test_df = pd.read_csv("https://raw.githubusercontent.com/skissler/haslemere/master/Kissler_DataS1.csv")
```

```python
test_df.info()
test_df.head()
```

```python
graph = nx.Graph()
for index, row in test_df.iterrows():
    graph.add_edge(row[1], row[2], distance=row[3], timestep=row[0])
print(len(graph.edges), len(graph.nodes))
nx.draw(graph, node_size=15, pos=nx.spring_layout(graph))
```

```python
nodes = list(graph.nodes())
edges = list(graph.edges(data=True))
print(nodes[0])
print(edges[0:10])
```
